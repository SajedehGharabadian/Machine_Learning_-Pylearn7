```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Dataset/Lionel Messi Goals.csv')



data
```

```python
data['Date'] = data['Date'].str[-5:]
data = data.rename(columns={"Goal Type": "Type"})
data = data.rename(columns={"Final Game Result": "Result"})
data = data.rename(columns={"Goal Method": "Method"})
data = data.rename(columns={"Score Home Team - When Messi Scored": "Score_home"})
data = data.rename(columns={"Score Away Team - When Messi Scored": "Score_away"})

data
```

```python
df = data['Type'].value_counts().rename_axis('Type').to_frame('counts').reset_index()

df
```

```python
plt.figure(figsize=(12,10))
plt.rcParams['font.size'] = 12.0
plt.pie(df['counts'], labels=df['Type'],autopct='%1.1f%%',frame=True)
plt.show()
```

```python
df = data['Method'].value_counts().rename_axis('Method').to_frame('counts').reset_index()

df
```

```python
plt.figure(figsize=(10,8))
plt.bar(df['Method'],df['counts'])
plt.show()
```

```python
data['Minute'] = data['Minute'].apply(eval)

data
```

```python
data['Minute'] = data['Minute'].astype(int)

df = data['Minute'].value_counts().rename_axis('Minute').to_frame('counts').reset_index().sort_values(by=['Minute'],ascending=True)

df
```

```python
plt.figure(figsize=(24,8))
plt.plot(df['Minute'],df['counts'])
plt.show()
```

```python
df = data.groupby(['Date'])['Score_away'].count().reset_index()

df
```

```python
plt.figure(figsize=(10,8))
plt.bar(df['Date'],df['Score_away'])
plt.show()
```

```python
df = data.query("Type == 'Penalty'")

df1 = df.groupby(['Method'])['Score_home','Score_away'].sum()


df1
```

```python
df = df1['Score_home'] + df1['Score_away']
Rightfoot = 0
df.loc['Rightfoot']= Rightfoot

df
```

```python
plt.figure(figsize=(8,8))
df.plot.bar(df,df)
plt.show()
```

```python
df = data.query("Minute > 90")

df2 = df.groupby(['Result'])['Score_home','Score_away'].sum()


df2
```

```python
df = df2['Score_home'] + df2['Score_away']

df
```

```python
plt.figure(figsize=(8,8))
df.plot.bar(df,df)
plt.show()
```
