```python
import numpy as np 
from numpy.linalg import inv
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.model_selection import train_test_split
```

```python
X = np.random.uniform(5,19,250)
Y = X + np.random.uniform(-10, 10, 250) + 2
```

```python
plt.scatter(X,Y)
```

```python
X = X.reshape((250,1))
Y = Y.reshape((250,1))
```

```python
m = np.matmul(inv(np.matmul(X.T,X)),np.matmul(X.T,Y))
m
```

```python
class LinearLeastSquare:
    def __init__(self):
        self.w = 0

    def fit(self, X_train , Y_train):
        self.w = np.matmul(np.matmul(np.linalg.inv( np.matmul(X_train.T  , X_train) ) , X_train.T ) , Y_train )

    def predict(self , X_test):
        y_pred = X_test @ self.w
        return y_pred
```

```python
lls = LinearLeastSquare()
lls.fit(X,Y)
y_pred = lls.predict(X)
```

```python
random_slope=np.random.uniform(0,4)
# random_slope
plt.scatter(X,Y)
plt.plot(X,random_slope*X,c="cyan",label="random slope")
plt.plot(X,y_pred,c='red',lw=4,label='fitted line')
plt.xlabel("study hours")
plt.ylabel("grade of students")
plt.legend()
plt.show()
```
