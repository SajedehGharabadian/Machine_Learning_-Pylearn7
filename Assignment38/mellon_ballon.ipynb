```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mellon = pd.DataFrame({
    'length': np.random.normal(25,1, 1000),
    'width': np.random.normal(13,1, 1000),
    'weight': np.random.normal(6,1,1000)}
)

ballon = pd.DataFrame({
    'length': np.random.normal(10, 1, 1000),
    'width': np.random.normal(4, 1, 1000),
    'weight': np.random.normal(2,1,1000)}
)

```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter3D(mellon['length'], mellon['width'],mellon['weight'],alpha = 0.5,c='yellow')
ax.scatter3D(ballon['length'], ballon['width'],ballon['weight'],alpha = 0.5,c='cyan')

ax.set_xlabel("width")
ax.set_ylabel("length")
ax.set_zlabel("weight")

plt.show()
```
